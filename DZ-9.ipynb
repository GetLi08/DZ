```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Students%20Performance.csv")
df.head()
```

# для получения суммы балов применяется apply

```python
df1 = df.copy()
summ = lambda x: x['math score']+x['reading score']+x['writing score']
df1['summary'] = df1.apply(summ, axis = 1)
df1.head()
```

# Использование map

```python
test_dict = {
  'none': '-',
    'completed': '+'
}
df['test preparation course'] = df['test preparation course'].map(test_dict)
df.head()
```

# Использование Transform

```python
df['mean_s'] = df.groupby('race/ethnicity')['math score'].transform('mean')
df.head()
```
