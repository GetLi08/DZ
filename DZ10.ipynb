```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Students%20Performance.csv")
df.head()
```

### Построение столбчатой диаграммы

```python
data = pd.DataFrame({"Дисциплина": ['math score','reading score','writing score'], "Средний бал": [df['math score'].mean(),df['reading score'].mean(),df['writing score'].mean()]})
sns.barplot(data = data, x = 'Дисциплина', y = 'Средний бал')
plt.show()
```

можно сделать вывод, что средняя успеваемость по трем показалям находится на высоком уровне и не имеет явного перекоса в сторону одной из дисциплин

### Построение гистограммы

```python
data = pd.DataFrame({'распределение math score': df['math score']})
sns.histplot(data, kde = True)
plt.show()
```

### Построение точечной диаграммы

```python
data = pd.DataFrame({"x": df['math score'], "y": df['writing score']})
sns.scatterplot(data = data, x = 'x', y = 'y')
plt.show()
```

По диаграмме визуально заметна некоторая зависимость показателя math score и writing score
Можно предположить, что студент имеющий высокий показатель math будет иметь высокий показатель writing

### Тепловая карта

```python
data2 = df.groupby(['gender','race/ethnicity'])['math score'].mean().unstack()
data2.head()
sns.heatmap(data2, annot = True, cmap="mako")
plt.show()
```

На тепловой карте явно заметно различие в среднем бале math score у представителей разных рас и пола
