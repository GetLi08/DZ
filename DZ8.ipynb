```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Students%20Performance.csv")
df.head()
```

Преобразование широкой формы в длинную

```python
dfcut = df.copy()
(
    dfcut
    .reset_index()
    .melt(
        id_vars = ['parental level of education','race/ethnicity','gender', 'lunch', 'test preparation course'], 
        var_name = 'score',
        value_name = 'val'
    )
)
```

```python
Использование pivot_table для оценки сренего значения math score в разрезе female/male в группах parental level of education
```

```python
dfcut.pivot_table(values = 'math score', index = 'parental level of education', columns = 'gender', aggfunc = 'mean' )
```

```python
С помощью groupby определим есть ли отличия в показателях writing score у студентов, походивших подготовку и не проходивших
```

```python
df.groupby('test preparation course')['writing score'].mean()
```

```python
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
```
